**Ejercicios de Limpieza**

1- Despu√©s de la uni√≥n de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df_union.drop(['country'], axis=1, inplace=True)
```

```python
df_union.head(1)
```

2- Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:   
    - No tengan espacios.   
    - Est√©n en min√∫scula.   
    - No tengan par√©ntesis, es decir, quitar "(%)", "(Km2)".   
    - Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.   
    - Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
#Que no tengan espacios y est√© en min√∫sculas

col_nuevas = {col : col.lower().replace(' ', '_') for col in df_union.columns}
df_union.rename(columns= col_nuevas, inplace= True)
```

```python
#Quitar par√©ntesis

col_parentesis = {col : col.replace('(_%)', '').replace('(%)', '').replace('(p/km2)', '').replace('(km2)', '') for col in df_union.columns}
df_union.rename(columns= col_parentesis, inplace= True)
```

```python
#Eliminar \n

col_n = {col : col.replace('\n', '') for col in df_union.columns}
df_union.rename(columns= col_n, inplace= True)
```

```python
#Eliminar :

col_puntos = {col : col.replace(':', '') for col in df_union.columns}
df_union.rename(columns= col_puntos, inplace= True)
```

```python
list(df_union.columns)
```

3- La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df_union['coordinates']
```

```python
df_union[['latitude', 'longitude']] = df_union['coordinates'].str.split(', ', n=1, expand=True)
```

```python
df_union.drop(['coordinates'], axis=1, inplace=True)
```

```python
df_union['latitude'] = df_union['latitude'].str.replace("'", "").str.replace(",", "").str.replace("(", "").str.replace(")", "").astype(float)


df_union['longitude'] = df_union['longitude'].str.replace("'", "").str.replace(",", "").str.replace("(", "").str.replace(")", "").astype(float)
```

```python
df_union.head(1)
```

4- Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
list(df_union.columns)
```

```python
columns_limpiar = ['unemployment_rate', 'total_tax_rate', 'out_of_pocket_health_expenditure','agricultural_land', 'cpi_change_', 'forested_area_', 'gross_primary_education_enrollment_', 'gross_tertiary_education_enrollment_', 'population_labor_force_participation_', 'tax_revenue_']
```

```python
for col in columns_limpiar:
    df_union[col] = df_union[col].str.replace('%', '')
```

```python
df_union
```

5- Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columnas_dolar = ['gasoline_price', 'gdp', 'minimum_wage']

for col in columnas_dolar:
    df_union[col] = df_union[col].str.replace('$', '')
```
